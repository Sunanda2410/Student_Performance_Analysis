# Student Performance Analysis

This notebook analyzes the Student Performance dataset using Python, Pandas, and Matplotlib. The goal is to clean, analyze, visualize, and draw conclusions from the data.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student-mat.csv", sep=";")

print(df.head())
```

## Data Cleaning and Exploration

In this step, we inspect the dataset shape, data types, missing values, and duplicate records.

```python
print("Dataset Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

## Analysis Questions

The following analysis answers key questions about student performance.

```python
# Average Final Grade
avg_grade = df["G3"].mean()
print("Average Final Grade (G3):", avg_grade)

# Students scoring above 15
above_15 = (df["G3"] > 15).sum()
print("Students scoring above 15:", above_15)

# Correlation
correlation = df["studytime"].corr(df["G3"])
print("Correlation between Study Time and G3:", correlation)

# Gender-wise performance
gender_avg = df.groupby("sex")["G3"].mean()
print("\nAverage Grade by Gender:")
print(gender_avg)
```

## Visualization 1: Distribution of Final Grades

This histogram shows how the final grades (G3) are distributed among students.

```python
plt.figure(figsize=(6,4))
plt.hist(df["G3"], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.show()
```

## Visualization 2: Study Time vs Final Grade

This scatter plot shows the relationship between study time and final grades.

```python
plt.figure(figsize=(6,4))
plt.scatter(df["studytime"], df["G3"])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
```

## Visualization 3: Average Grade by Gender

This bar chart compares the average final grade of male and female students.

```python
plt.figure(figsize=(6,4))
gender_avg.plot(kind="bar")
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```

# Conclusion

- The dataset contains 395 students and 33 attributes.
- No missing values were found in the dataset.
- No duplicate records were found.
- The average final grade (G3) is approximately 10.42.
- 40 students scored above 15 in the final grade.
- Study time has a weak positive correlation with final grade (0.098).
- Male students achieved a slightly higher average final grade than female students.
- Visualizations helped understand grade distribution, study habits, and gender-wise performance.
